```python
#Making a copy of the dataframe
bike_data_copy = bike_data[['ride_id','rideable_type','started_at','ended_at','start_station_name','end_station_name','start_lat','start_lng','end_lat','end_lng','member_casual']].copy()
bike_data_copy.head()
```

```python
bike_data_copy['rideable_type'].unique()
```

```python
#Investigating the rideable_type on Divvy bikes website, we understand that the docked_bike and classic_bike are used interchangeably.
bike_data_copy['rideable_type']=bike_data_copy['rideable_type'].replace('docked_bike','classic_bike')
```

```python
bike_data_copy['rideable_type'].unique()
```

```python
# Start and end times are in object format and different formats of time. Converting them into date_time
bike_data_copy['started_at']=pd.to_datetime(bike_data_copy['started_at'])
bike_data_copy['ended_at']=pd.to_datetime(bike_data_copy['ended_at'])
bike_data_copy.info()
```

```python
bike_data_copy['start_station_name'].value_counts()
```

```python
# Erase (*) and (Temp) at the end of station name
bike_data_copy['start_station_name'] = bike_data_copy['start_station_name'].str.rstrip('(*) (Temp)')
bike_data_copy['end_station_name'] = bike_data_copy['end_station_name'].str.rstrip('(*) (Temp)')
```

```python
bike_data_copy.shape
```

```python
#Filling missing or Null values with "Unknown" for start_station_name and end_station_name
bike_data_copy= bike_data_copy.fillna('Unknown')
bike_data_copy.isnull().sum()
```

```python
# ride_duration column(ended_at-started_at) changed to minutes
bike_data_copy['ride_duration'] = (bike_data_copy['ended_at']-bike_data_copy['started_at']).dt.total_seconds()/60                               

print("Minimum ride_duration : ", bike_data_copy['ride_duration'].min())
print("Maximum ride_duration : ", bike_data_copy['ride_duration'].max())
```

```python
# Remove data with ride duration<0
bike_data_copy = bike_data_copy[bike_data_copy['ride_duration'] > 0]
bike_data_copy.info()
```

```python
bike_data_copy[['start_station_name','end_station_name','ride_duration']]
```

```python
#If start and end station name are same and ride_duration <1 minute, we can assume the trip did not happen. Maybe bike was returned.
Not_confirmed_trip= bike_data_copy.loc[(bike_data_copy['start_station_name'] == bike_data_copy['end_station_name']) & (bike_data_copy['ride_duration']<1)].copy()
Not_confirmed_trip
```

```python
idx = Not_confirmed_trip.index.to_list()
bike_data_copy = bike_data_copy.drop(idx)
bike_data_copy.head()
```

```python
bike_data_copy.shape
```

```python
#Renaming column for better understanding
bike_data_copy = bike_data_copy.rename(columns={'member_casual':'customer_type'})
bike_data_copy.head()
```

```python
# Finding any outliers
Q1 = bike_data_copy['ride_duration'].quantile(0.25)
Q3 = bike_data_copy['ride_duration'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5*IQR
lower_bound = Q1 - 1.5*IQR
print("Upper_bound and Lower bound: \n",upper_bound, lower_bound)
print("Number of data points greater than upper_bound: \n",bike_data_copy[(bike_data_copy['ride_duration'] > (upper_bound))].count())
print("Number of data points lower than lower_bound: \n",bike_data_copy[(bike_data_copy['ride_duration'] < (lower_bound))].count())
```

```python
#For Divvy bikes, the maximum allowed ride duration varies depending on the type of membership and bike.
#But ideally it doesn't exceed 3 hours(90 minutes). Hence i shall consider data above it as outliers.
#In my understanding, the upper bound specified by IQR is low and shall do overfitting. 
#Lower bound is in negative value, hence it doesn't really matter as we have already removed data with ride_duration<0
bike_data_copy[(bike_data_copy['ride_duration'] > 90)].count()
```

```python
#Removing the outliers
bike_data_copy = bike_data_copy[bike_data_copy['ride_duration'] <= 90]
```

Important Visualizations:
Plot-1: Number of rides per day throughout years 2022-2023

```python
data_2022 = bike_data_copy[bike_data_copy['started_at'].dt.year == 2022]
data_2023 = bike_data_copy[bike_data_copy['started_at'].dt.year == 2023]

# Group by date to get daily ride counts
daily_rides_2022 = data_2022.groupby(data_2022['started_at'].dt.date).size()
daily_rides_2023 = data_2023.groupby(data_2023['started_at'].dt.date).size()

# Calculate mean values
mean_2022 = daily_rides_2022.mean()
mean_2023 = daily_rides_2023.mean()

# Plot daily rides for each year with mean lines
plt.figure(figsize=(14, 6))

# Plot 2022 data
sns.lineplot(x=daily_rides_2022.index, y=daily_rides_2022.values, color="blue", label='2022')
plt.axhline(mean_2022, color='blue', linestyle='--', label=f'2022 Mean ({mean_2022:.1f})')

# Plot 2023 data
sns.lineplot(x=daily_rides_2023.index, y=daily_rides_2023.values, color="green", label='2023')
plt.axhline(mean_2023, color='green', linestyle='--', label=f'2023 Mean ({mean_2023:.1f})')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Number of Rides')
plt.title('Number of Bike Rides per Day (2022 and 2023)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Plot-2: Number of rides every day of week throughout years 2022-2023

```python
from matplotlib.ticker import ScalarFormatter
# Extract day of the week
bike_data_copy['day_of_week'] = bike_data_copy['started_at'].dt.day_name()

# Group by day of the week and count rides
rides_by_day = bike_data_copy.groupby('day_of_week').size().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Calculate the mean for the week
weekly_mean = rides_by_day.mean()

#Plot figure
plt.figure(figsize=(10, 5))
sns.lineplot(x=rides_by_day.index, y=rides_by_day.values, marker='o', color='purple', label='Rides per Day')
plt.axhline(weekly_mean, color='purple', linestyle='--', label=f'Mean ({weekly_mean:.1f})')

# Add labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Number of Rides')
plt.title('Number of Bike Rides by Day of the Week')
plt.legend()
plt.show()
```

Plot-3: Number of trips starting from a station(based on latitude and longitude)

```python
# Plot start stations by latitude and longitude
plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=bike_data_copy, x='start_lat', y='start_lng', 
    hue='start_station_name', palette="tab10", legend=False, s=20
)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Start Stations by Latitude and Longitude')
plt.show()
```

Plot-4: Number of rides per hour

```python
# Extract the hour from the 'start_time'
bike_data_copy['hour'] = bike_data_copy['started_at'].dt.hour

# Group by hour and count the number of rides
rides_per_hour = bike_data_copy.groupby('hour').size()
#Mean
hourly_mean = rides_per_hour.mean()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(x=rides_per_hour.index, y=rides_per_hour.values, marker='o', color='purple', label='Rides per hour')
plt.axhline(hourly_mean, color='purple', linestyle='--', label=f'Mean ({hourly_mean:.1f})')

plt.xlabel('Hour of the Day')
plt.ylabel('Number of Rides')
plt.title('Number of Bike Rides by hour of the Day')
plt.legend()
plt.show()
```
